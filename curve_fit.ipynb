# curve fitting

```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.mlab as mlab
from scipy import optimize, polyfit
```

```python
n = 50
jitter_amp = 4.0
x = np.linspace(0,10, n)
jitter = jitter_amp * (np.random.random(n) - .5)
y = x + jitter
plt.plot(x,y, 'o')
```

```python
m, b = polyfit(x,y,1)
```

```python
(m,b)
```

```python
plt.plot(x,y, 'o')
t = np.array([0,10])
plt.plot(t, t*m+b)
```

```python
def fit_func(t,m,b):
    return t*m+b
```

```python
p, cov = optimize.curve_fit(fit_func, x, y)
```

```python
p
```

```python
plt.plot(x,y, 'o')
t1 = np.array([0,10])
plt.plot(t1, t1*p[0]+p[1])
```
